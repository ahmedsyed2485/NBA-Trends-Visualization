```python
all_stats = all_stats.fillna(0)
all_stats = all_stats.loc[all_stats['Year'] > 1955]


all_stats.head()
```

```python
all_stats['PPG'] = all_stats['PTS'] / all_stats['G']
all_stats['APG'] = all_stats['AST'] / all_stats['G']
all_stats['RPG'] = all_stats['TRB'] / all_stats['G']
all_stats['2PAPG'] = all_stats['2PA'] / all_stats['G']
all_stats['FGAPG'] = all_stats['FGA'] / all_stats['G']
all_stats['3PAPG'] = all_stats['3PA'] / all_stats['G']
all_stats['MPPG'] = all_stats['MP'] / all_stats['G']

all_stats.head()
```

```python
all_stats = all_stats[['Year', 'Player', 'Pos', 'G', 'MP', '2PAPG', '3PA', 'FGA', 'eFG%','3PAPG','FGAPG','MPPG','APG', 'PPG', 'PTS']]
all_stats.head()
```

```python
all_stats = all_stats.loc[(all_stats['Year'] > 1955) & (all_stats['G'] > 24) & (all_stats['MPPG'] > 20)]
all_stats.head()
```

```python
C = round(all_stats.loc[all_stats['Pos'] == 'C'].groupby('Year').mean(),2)
PF = round(all_stats.loc[all_stats['Pos'] == 'PF'].groupby('Year').mean(),2)
SF = round(all_stats.loc[all_stats['Pos'] == 'SF'].groupby('Year').mean(),2)
SG = round(all_stats.loc[all_stats['Pos'] == 'SG'].groupby('Year').mean(),2)
PG = round(all_stats.loc[all_stats['Pos'] == 'PG'].groupby('Year').mean(),2)
```

```python
years_shotsgame = C.index
```

```python

layout = go.Layout(
    title = 'NBA POSITIONS FGA (1955-2017)',
    yaxis = dict(
        title = 'Average FGA'
    ),
     xaxis = dict(
         title = 'Years'
    )
     
)

pg = go.Scatter( x=years_shotsgame, y=PG['FGAPG'], name='Point Guards' , line= dict(color='blue'))
sg = go.Scatter( x=years_shotsgame, y=SG['FGAPG'], name='Shootings Guards' , line= dict(color='red'))
sf = go.Scatter( x=years_shotsgame, y=SF['FGAPG'], name='Small Forwards' , line= dict(color='yellow'))
pf = go.Scatter( x=years_shotsgame, y=PF['FGAPG'], name='Power Forwards' , line= dict(color='orange'))
c = go.Scatter( x=years_shotsgame, y=C['FGAPG'], name='Centers' , line= dict(color='purple'))
            
Test1 = [pg,sg,sf,pf,c]


fig = go.Figure(data=Test1, layout=layout)
fig.show(renderer ='svg')
```
